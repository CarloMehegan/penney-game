```python
import numpy as np
import pandas as pd
import random
import itertools 
import os
```

```python
#load in list of simulations, convert ot binary
def load_process_simulations(file_path): 
    simulations = np.load(file_path)
    
    def int_to_bit(n):
        binary = bin(n)[2:] 
        return binary.zfill(52)
    
    binary_simulations = [int_to_bit(sim) for sim in simulations]             
    return binary_simulations
```

```python
def variation1(deck, player1_sequence, player2_sequence):
    player1_cards = 0
    player2_cards = 0
    pile = 0
    
    for i in range(len(deck) - 2):
        current_sequence = deck[i:i+3]
        pile += 1
        if current_sequence == player1_sequence:
            player1_cards += pile
            pile = 0
        elif current_sequence == player2_sequence:
            player2_cards += pile
            pile = 0 
    return deck, player1_cards, player2_cards
```

```python
def process_decks_v1(processed_sims):
    results = []  # Initialize results list

    # Loop through each deck in processed_sims
    for deck in processed_sims:
        all_sequences = [''.join(seq) for seq in itertools.product('01', repeat=3)]
        
        # Loop through all pairs of sequences
        for p1 in all_sequences:
            for p2 in all_sequences:
                if p1 != p2:
                    deck, p1_cards, p2_cards = variation1(deck, p1, p2)
                    seq_combo = f"{p1}, {p2}"

                    # Append results for the current combination
                    results.append({
                        'Deck': deck,
                        'Sequence Combo': seq_combo,
                        'Player 1 Cards': p1_cards,
                        'Player 2 Cards': p2_cards
                    })

    # Convert results to DataFrame
    df = pd.DataFrame(results)
    return df
```

```python
def variation2(deck, player1_sequence, player2_sequence):
    player1_tricks = 0
    player2_tricks = 0
    
    for i in range(len(deck) - 2):
        current_sequence = deck[i:i+3]
        if current_sequence == player1_sequence:
            player1_tricks += 1
        elif current_sequence == player2_sequence:
            player2_tricks += 1
    return deck, player1_tricks, player2_tricks
```

```python
def process_decks_v2(processed_sims):
    results = []  # Initialize results list

    # Loop through each deck in processed_sims
    for deck in processed_sims:
        all_sequences = [''.join(seq) for seq in itertools.product('01', repeat=3)]
        
        # Loop through all pairs of sequences
        for p1 in all_sequences:
            for p2 in all_sequences:
                if p1 != p2:
                    deck, p1_tricks, p2_tricks = variation2(deck, p1, p2)
                    seq_combo = f"{p1}, {p2}"

                    # Append results for the current combination
                    results.append({
                        'Deck': deck,
                        'Sequence Combo': seq_combo,
                        'Player 1 Tricks': p1_tricks,
                        'Player 2 Tricks': p2_tricks
                    })

    # Convert results to DataFrame
    df = pd.DataFrame(results)
    return df
```
